```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
n=100
x, y = make_regression(n_samples=n, n_features=1, noise=20)
x = x.reshape(x.shape[0])
y = y.reshape(x.shape[0])
plt.scatter(x,y)
```

### $$a=\frac{\overline{xy}-\overline{x}\ \overline{y}}{\overline{x^2}-\overline{x}^2}$$ $$b=\overline{y}-a\overline{x}$$

```python
a = (np.mean(x*y)-np.mean(x)*np.mean(y))/(np.mean(x*x)-np.mean(x)*np.mean(x))
b = np.mean(y)-a*np.mean(x)

print("a :",a)
print("b :",b)
```

```python
plt.scatter(x,y)
plt.plot((3,-3),(a*3+b,a*-3+b),c='r')
```
